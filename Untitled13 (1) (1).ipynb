```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mars= pd.read_excel("/content/MARS evaluation.xlsx")
mars.head()
```

```python
mars.info()
```

```python
#checking for missing values
mars.isnull().sum()
```

```python
mars.describe()
```

```python
mars.head()


```

```python
# List of columns you want to drop
columns_to_drop = ['credibility', 'evidence base', 'Engagement', 'Functionality', 'Aesthetics',  'MARS score']

# Create mars1 DataFrame by dropping the specified columns
mars1 = mars.drop(columns=columns_to_drop)

# Display mars1 DataFrame
print(mars1.head())
```

```python
mars1.head()
```

```python
mars1=mars1.drop(columns=['Platform'], axis=1)
```

```python
mars1.head()
```

```python
# Drop the last column from mars1 DataFrame
mars1 = mars1.iloc[:, :-1]

# Display the updated mars1 DataFrame
mars1.head()
```

```python
# @title Indivudual adaptability

from matplotlib import pyplot as plt
mars1['Indivudual adaptability'].plot(kind='hist', bins=20, title='Indivudual adaptability')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# heatmap
correlation=mars1.corr()
plt.figure(figsize=(8,8))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Greens')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming mars is already loaded as a DataFrame

# List of columns to plot
columns_to_plot = ['Engagement', 'Functionality', 'Aesthetics','Information quality']

# Create box plots
mars[columns_to_plot].boxplot()

# Set plot title and labels
plt.title('Box Plots of Mars Data')
plt.xlabel('Attributes')
plt.ylabel('Score')

# Show plot
plt.show()
```
